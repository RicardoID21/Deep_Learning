# Algorithmic Trading Strategy — Performance Report

This notebook presents the design, implementation, and evaluation of a rule-based trading algorithm optimized using Optuna. The strategy operates on 5-minute historical data for AAPL and employs technical indicators (EMA, SMA, Bollinger Bands) combined with a signal aggregation method.


## Strategy Design

**Indicators used:**
  - Exponential Moving Average (EMA)
  - Simple Moving Average (SMA)
  - Bollinger Bands (Upper/Lower)
  
**Signal Combination:**
  - Each indicator emits a signal: +1 (buy), -1 (sell), 0 (neutral)
  - A combined signal (`Cash_medio`) is generated only when at least **2 out of 3** indicators agree
